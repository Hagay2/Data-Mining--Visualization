```python
#Using pandas for loading the csv file into a dataframe
import pandas as pd

#Using matplotlib library for basic graph plotting
import matplotlib.pyplot as plt

#Using seaborn library that allows us to optimize matplotlib's output
import seaborn as sns
```

```python
#Loading the csv file into the pandas dataframe
```

```python
wines= pd.read_csv('WineQT.csv')
wines
```

```python
#Presenting the first 10 rows 
```

```python
wines.head(10)
```

```python
#Generating line plot
```

```python
wines.plot(kind='line', x = 'chlorides', y = 'alcohol')
```

```python
#Generating scatter plot
```

```python
wines.plot(kind='scatter', x = 'fixed acidity', y = 'citric acid')
```

```python
#Generating histogram plot
```

```python
wines.plot(kind='hist', x = 'fixed acidity', y = 'total sulfur dioxide')
```

```python
#Generating Kernel Density Estimation plot
```

```python
wines.plot(kind='kde', x = 'alcohol', y = 'volatile acidity')
```

```python
#Generating area plot
```

```python
wines.plot(kind='area', x = 'fixed acidity', y = 'sulphates')
```

```python
#Using seaborn to create a joinplot
```

```python
sns.jointplot(x = 'alcohol', y = 'quality', data = wines)
```

```python
#Using seaborn to create a relplot of the first 30 rows
```

```python
sns.relplot(data=wines.head(30), x="total sulfur dioxide", y="pH")
```

```python
#Using seaborn to create a stripplot of the first 30 rows
```

```python
sns.stripplot(x="residual sugar", y="quality", data=wines.head(30))
```

```python
#Generating bar plot of the first 30 rows
```

```python
wines.head(30).plot(kind='bar', x = 'citric acid', y = 'chlorides')
```

```python
#Using seaborn to create jointplot of Kernel Density Estimation of the first 30 rows
```

```python
sns.jointplot(data=wines.head(30), x="chlorides", y="citric acid", kind="kde")
```
